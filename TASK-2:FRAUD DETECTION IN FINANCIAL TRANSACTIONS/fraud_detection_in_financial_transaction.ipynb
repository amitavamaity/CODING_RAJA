```python
legit_df = df[df["Class"] == 0]
illegal_df = df[df["Class"] == 1]
```

```python
legit_df.head()
```

```python
illegal_df.head()
```

```python
df.groupby("Class").mean()
```

```python
legit_rnd_df = legit_df.sample(1500)
legit_rnd_df.shape
```

```python
new_df = pd.concat([legit_rnd_df, illegal_df])
```

```python
new_df.groupby("Class").mean()
```

# Data splting, training, testing


```python
x = new_df.iloc[:, :30]
x.head()
```

```python
y = new_df.iloc[: , 30]
y.head()
```

```python
x = x.values
y = y.values
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.1, random_state = 5)
```

```python
reg = LogisticRegression()
```

```python
reg.fit(x_train, y_train)
```

```python
y_pred = reg.predict(x_test)
```

```python
reg.predict([input1])
```

```python
accuracy_score(y_pred, y_test)
```
